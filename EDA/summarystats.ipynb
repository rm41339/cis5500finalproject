```python
import kagglehub
import pandas as pd
```

```python
# Download latest version, replace with dataset you want to analyze
path = kagglehub.dataset_download("maxhorowitz/nflplaybyplay2009to2016")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(path + "/playbyplay2018.csv")
```

```python
for col in df.columns:
    print(f"Column: {col}")
    print(f"  Missing values: {df[col].isnull().sum()}")
    if pd.api.types.is_numeric_dtype(df[col]):
        print(f"  Mean: {df[col].mean()}")
        print(f"  Median: {df[col].median()}")
        print(f"  Std Dev: {df[col].std()}")
    else:
        print(f"  Unique values: {df[col].nunique()}")
        print(f"  Top 5 frequent values:\n{df[col].value_counts().head()}")
    print("\n")
```
